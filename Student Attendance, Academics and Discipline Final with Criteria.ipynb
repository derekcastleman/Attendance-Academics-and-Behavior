## Attendance

Attendance will be looked at in three different ways. First, the number of periods that a student was tardy to class will be summed up. I will follow this by calculating the number of periods that each student missed due to an unexcused absence. It will conclude by calculating the total periods that a student missed whether or not they were excused or unexcused.

```python
codes
```

```python
codes['Date']= pd.to_datetime(codes['Date'])
codes
```

```python
a = input('What is the start date you are interested in (mm/dd/yyyy):          ')
a
```

```python
b = input('What is the end date you are interested in (mm/dd/yyyy):          ')
b
```

```python
codes = codes[(codes['Date'] >= a) & (codes['Date'] <= b)]
codes
```

```python
codes = codes.melt(['School','Student#', 'Student ID', 'Date'], var_name='Period', value_name='Code')
codes
```

```python
totalabsent = codes[(codes['Code'] == 'A') | (codes['Code'] == 'Q') | 
                     (codes['Code'] == 'S') | (codes['Code'] == 'U') | (codes['Code'] == 'K') |
                      (codes['Code'] == 'I') | (codes['Code'] == 'M') | (codes['Code'] == 'X')]
totalabsent
```

```python
totalabsent['Periods Absent'] = 1
totalabsent
```

```python
totalabsent = totalabsent.groupby(by=['Student ID'])['Periods Absent'].sum().reset_index()
totalabsent
```

```python
unexcused = codes[(codes['Code'] == 'A') | (codes['Code'] == 'Q') | 
                     (codes['Code'] == 'S') | (codes['Code'] == 'U')]
unexcused
```

```python
unexcused['Unexcused Absent'] = 1
unexcused
```

```python
unexcused = unexcused.groupby(by=['Student ID'])['Unexcused Absent'].sum().reset_index()
unexcused
```

```python
totaltardy = codes[(codes['Code'] == 'C') | (codes['Code'] == 'D') | 
                     (codes['Code'] == 'T') | (codes['Code'] == 'Y') | (codes['Code'] == 'Z')]
totaltardy
```

```python
totaltardy['Periods Tardy'] = 1
totaltardy
```

```python
totaltardy = totaltardy.groupby(by=['Student ID'])['Periods Tardy'].sum().reset_index()
totaltardy
```

```python
attendance = pd.merge(totalabsent, unexcused, how='outer', on='Student ID')
attendance
```

```python
attendance = pd.merge(attendance, totaltardy, how='outer', on='Student ID')
attendance
```

## Discipline

A date will be input that will be selected for when discipline should be looked at beyond. A number of one will be assigned to each discipline log and then it will be summed for each of the students. The same thing will be done with the suspensions for a student.

```python
discipline
```

```python
discipline['Date']= pd.to_datetime(discipline['Date']) # Changing to proper date format
discipline
```

```python
discipline = discipline[(discipline['Date'] >= a) & (discipline['Date'] <= b)]
discipline
```

```python
discipline['Discipline Logs'] = 1 # Giving a value of one for each discipline
discipline
```

```python
# Summing up the discipline for each student
discipline = discipline.groupby(by=['Student ID'])['Discipline Logs'].sum().reset_index()
discipline
```

```python
suspension
```

```python
suspension = suspension[(suspension['Disposition'] == 'SUS-13') | (suspension['Disposition'] == '04')] # Selecting suspensions
suspension
```

```python
# Changing date into proper format
suspension['Disposition date']= pd.to_datetime(suspension['Disposition date'])
suspension
```

```python
# Selecting suspensions beyond the start 
suspension = suspension[(suspension['Disposition date'] >= a) & (suspension['Disposition date'] <= b)]
suspension
```

```python
suspension['Suspensions'] = 1 # Giving a value of 1 for each suspension
suspension
```

```python
# Summing up the suspensions for each student
suspension = suspension.groupby(by=['Student ID'])['Suspensions'].sum().reset_index()
suspension
```

```python
# Merging the two behavior columns together
behavior = pd.merge(discipline, suspension, how='outer', on='Student ID')
behavior
```

```python
behavior = behavior.fillna(0) # Filling in the missing values with zero
behavior
```

The fees will be calculated for students by subtracting the amount paid from the amount charged. The community service hours will then be calculated by filtering over the date range and then summing the column for the hours. Everything will then be added to the dataframe to complete it.

```python
fees
```

## Criteria

An input of various criteria is asked for and then students are determined if they meet all criteria or how many they are missing which is put in one final column.

```python
# Input period tardies allowed
m = input('What is the maximum for period tardies allowed:          ')
m = int(m)
m
```

```python
# Input period absences allowed
n = input('What is the maximum for absences allowed:          ')
n = int(n)
n
```

```python
# Input discipline allowed
o = input('What is the maximum for discipline logs allowed:          ')
o = int(o)
o
```

```python
# Input suspensions allowed
p = input('What is the maximum for suspensions allowed:          ')
p = int(p)
p
```

```python
# Input fees allowed
q = input('What is the maximum for fees allowed:          ')
q = int(q)
q
```

```python
# Input gpa needed
r = input('What is the minimum gpa:          ')
r = int(r)
r
```

```python
# Input D/F allowed
s = input('What is the maximum number of D/F allowed:          ')
s = int(s)
s
```

```python
# Input community service hours needed
t = input('What is the minimum hours of community service:          ')
t = int(t)
t
```

```python
final
```

```python
final
```

```python
final = final.drop_duplicates(subset=['Student ID'], keep='first')
final
```

```python
# Selects students who meet all criteria
qualify = final[(final['Periods Absent'] <= n) & (final['Periods Tardy'] <= m) & (final['Discipline Logs'] <= o) & 
               (final['Suspensions'] <= p) & (final['Q3 D/F'] <= s) & (final['Fees/Debts'] <= q) & (final['TGPA'] >= r) &
               (final['Hours'] >= t)]
qualify
```

```python
# Creates a column that says they qualify
qualify['Qualify/#Off'] = 'Yes'
qualify
```

```python
# Chooses the students who do not qualify
unqualify = final[(final['Periods Absent'] > n) | (final['Periods Tardy'] > m) | (final['Discipline Logs'] > o) | 
               (final['Suspensions'] > p) | (final['Q3 D/F'] > s) | (final['Fees/Debts'] > q) |(final['TGPA'] < r) |
               (final['Hours'] < t)]
unqualify
```

```python
# Gives one for students with too many absents
period = unqualify[unqualify['Periods Absent'] > n]
period['Qualify/#Off'] = 1
period
```

```python
# Gives one for students with too many tardies
tardy = unqualify[unqualify['Periods Tardy'] > m]
tardy['Qualify/#Off'] = 1
tardy
```

```python
# Gives one for students with too many discipline
discipline = unqualify[unqualify['Discipline Logs'] > o]
discipline['Qualify/#Off'] = 1
discipline
```

```python
# Gives one for students with too many suspensions
suspensions = unqualify[unqualify['Suspensions'] > p]
suspensions['Qualify/#Off'] = 1
suspensions
```

```python
# Gives one for students with too many fees
fines = unqualify[unqualify['Fees/Debts'] > q]
fines['Qualify/#Off'] = 1
fines
```

```python
# Gives one for students with too many D/Fs
grades = unqualify[unqualify['Q3 D/F'] > s]
grades['Qualify/#Off'] = 1
grades
```

```python
# Gives one for students with too low of GPA
gpa = unqualify[unqualify['TGPA'] < r]
gpa['Qualify/#Off'] = 1
gpa
```

```python
# Gives one for students with too low of hours
hours = unqualify[unqualify['Hours'] < t]
hours['Qualify/#Off'] = 1
hours
```

```python
# Puts all the unqualifying groups together
under = pd.concat([period, tardy, discipline, suspensions, fines, grades, gpa, hours])
under
```

```python
# Groups by Student ID and adds how many criteria they did not meet
under = under.groupby(['Student ID'])['Qualify/#Off'].sum().reset_index()
under
```

```python
unqualify
```

```python
# Combines with the unqualify dataframe
unqualify = pd.merge(unqualify, under, how='inner', on='Student ID')
unqualify
```

```python
# Puts qualify and unaqualify back together
final = pd.concat([qualify, unqualify])
final
```

```python
# Renames the hours column
final = final.rename(columns={'Hours': 'Community Service'})
final
```
